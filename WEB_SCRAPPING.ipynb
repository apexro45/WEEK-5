```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://books.toscrape.com/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    print("Web page successfully loaded.")
else:
    print("Failed to load the web page.")
```

```python
# Extract book titles
book_titles = soup.find_all('h3')

for title in book_titles:
    print(title.get_text())
```

```python
import pandas as pd

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')


# Extract book information
books_data = []

for book in soup.find_all('article', class_='product_pod'):
    title = book.h3.a.attrs['title']
    rating = book.p.attrs['class'][1]
    price = book.select('div p.price_color')[0].get_text()
    link = url + book.h3.a.attrs['href']

    books_data.append({
        'Title': title,
        'Rating': rating,
        'Price': price,
        'Link': link
    })
```

Creating a datafram to display it in tabular form

```python
books_df = pd.DataFrame(books_data)
```
